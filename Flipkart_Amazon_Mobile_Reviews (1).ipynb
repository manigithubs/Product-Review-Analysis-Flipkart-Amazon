```python
df1.reset_index(inplace = True,drop = True)
def spell(text):
  txt = TextBlob(text)
  txt = txt.correct()
  return txt

def convert_emoji(text):
  replace_emoji = emoji.demojize(text)
  return replace_emoji.replace(':','').replace('_',' ')
```

```python

stop_words = set(stopwords.words("english"))
tokenizer = RegexpTokenizer('\w+')
for i in range(len(df1.axes[0])):
  z1=df1['Review-Body'][i]
  z=convert_emoji(z1)
  y=tokenizer.tokenize(z)
  filtered_list = [word.lower() for word in y if word.casefold() not in stop_words and len(word)>2 and len(word)<11]
  singular1 = [singularize(word) for word in filtered_list]
  le1 = [lemmatizer.lemmatize(word) for (word, pos) in nltk.pos_tag(singular1) if word.isalpha() and (pos[0]=='N' or pos[0]=='J' or pos[0]=='V' or pos[0]=='R')]
  le = [spell(word) for word in le1]
  if any(c in le for c in ['camera','picture','pic','image','video', 'selfie']):
    df1['camera'][i]=1
  if any(c in le for c in ['battery', 'day', 'charging', 'charge', 'day', 'drain','dead']):
    df1['battery'][i]=1
  if any(c in le for c in ['display', 'refresh', 'beautiful', 'pixel', 'screen','touch', 'response','look']):
    df1['display'][i]=1
  if any(c in le for c in ['value', 'money', 'cheap' , 'better' , 'best', 'segment' ]):
    df1['value_for_money'][i]=1
  if any(c in le for c in ['performance', 'hang', 'slow', 'smooth', 'glitch', 'UI', 'lagg']):
    df1['performance'][i]=1
  a=' '.join(str(e) for e in le)
  df1['Review-Body'][i]=a
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer

df2['Sentiment'] = None
df2['camera_s'] = None
df2['battery_s'] = None
df2['display_s'] = None
df2['value_for_money_s'] = None
df2['performance_s'] = None

def polar(a,b,c):
  if a>=0:
    df2.at[b,c]=1
  else:
    df2.at[b,c]=0

def get_context(string, target_word):
  try:
    words = string.split()
    target_index = words.index(target_word)
    before = words[max(target_index - 3, 0):target_index]
    after = words[target_index + 1:target_index + 4]
    if len(before) < 3:
        before.extend(words[0:3-len(before)])
    if len(after) < 3:
        after.extend(words[-3+len(after):])
    context = " ".join(before + [target_word] + after)
    return context
  except ValueError:
    return 'good'

sia = SentimentIntensityAnalyzer()

for index, row in df2.iterrows():
  polarity = sia.polarity_scores(row['Review-Body'])
  polar(polarity['compound'],index,'Sentiment')

  if df2['camera'][index]==1:
    for word in camera:
      z=df2['Review-Body'][index]
      if word in z:
        text=get_context(z,word)
        polarity = sia.polarity_scores(text)
        polar(polarity['compound'],index,'camera_s')
        break
  if df2['battery'][index]==1:
    for word in battery:
        z=df2['Review-Body'][index]
        if word in z:
          text=get_context(z,word)
          polarity = sia.polarity_scores(text)
          polar(polarity['compound'],index,'battery_s')
          break
  if df2['display'][index]==1:
    for word in display:
      z=df2['Review-Body'][index]
      if word in z:
        text=get_context(z,word)
        polarity = sia.polarity_scores(text)
        polar(polarity['compound'],index,'display_s')
        break
  if df2['value_for_money'][index]==1:
    for word in value_for_money:
      z=df2['Review-Body'][index]
      if word in z:
        text=get_context(z,word)
        polarity = sia.polarity_scores(text)
        polar(polarity['compound'],index,'value_for_money_s')
        break
  if df2['performance'][index]==1:
    for word in performance:
      z=df2['Review-Body'][index]
      if word in z:
        text=get_context(z,word)
        polarity = sia.polarity_scores(text)
        polar(polarity['compound'],index,'performance_s')
        break

df2=df2.fillna(np.nan)
```
